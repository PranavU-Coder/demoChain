```python
from typing import *
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values: List[int]
    name: str
    operation: str
    result: str
```

```python
def process(state: AgentState) -> AgentState:
    if state['operation'] == "+":
        state['result'] = sum(state['values'])
    elif state['operation'] == "*":
        state['result'] = 1  
        for val in state['values']:  
            state['result'] *= val
    else:
        print("Invalid operator, please use + or *")
        state['result'] = "0"  
    
    state['result'] = f"hi {state['name']} today your total is: {state['result']}"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("processor",process)

graph.set_entry_point("processor")
graph.set_finish_point("processor")

app = graph.compile()
```

```python
result = app.invoke({"values": [1,2,3,4,5], "name": "pranav", "operation": "*"})
print(result["result"])
```
